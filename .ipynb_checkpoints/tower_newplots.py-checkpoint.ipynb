```python
# Part 2: Theme Park Data Analysis
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate #This isn't working -- ImportError: no module named scipy
%matplotlib inline

# Loads the data
vdata=np.loadtxt("droptower_vdata.txt", dtype=float)

## Individual plots of displacement, velocity, acceleration, with labels, titles, line styles
# Plot position over time
x=integrate.cumtrapz(vdata[:,1], vdata[:,0], initial=0)
xplot=plt.plot(vdata[:,0], x)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.suptitle('Position vs. time',fontsize=18)
#xplot.savefig('xplot.png',dpi=200) <-- Uncomment to save plot
plt.show()

# Plot velocity over time
vplot=plt.plot(vdata[:,0], vdata[:,1])
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.suptitle('Velocity vs. time',fontsize=18)
#vplot.savefig('vplot.png',dpi=200)
plt.show()

# Plot acceleration over time
a=np.diff(vdata[:,1])
aplot=plt.plot(vdata[1:,0],a)
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s^2)')
plt.suptitle('Acceleration vs. time',fontsize=18)
#aplot.savefig('aplot.png',dpi=200)
plt.show()

## Plots of all 3 lines on the same pair of axes (should be distinguishable and have a legend)
allplot=plt.plot(vdata[:,0], x, color='red',linestyle='dotted', marker='o', label='Position') # position
plt.plot(vdata[:,0], vdata[:,1], color='blue',linestyle='dashed',marker='o', label='Velocity') # velocity
plt.plot(vdata[1:,0],a, color='green',linestyle='dashed', marker='o', label='Acceleration')
plt.xlabel('Time (s)')
plt.legend()
plt.suptitle('Drop tower data',fontsize=18)
#allplot.savefig('allplot.png', dpi=200)
plt.show()

## All 3 as subplots in the same figure
f, (ax, av, aa) = plt.subplots(3, sharex=True)
ax.plot(vdata[:,0], x,color='r', label='Position')
ax.set_xlim([0,10])
av.plot(vdata[:,0], vdata[:,1],color='b',label='Velocity')
aa.plot(vdata[1:,0],a,color='g', label='Acceleration')
plt.xlabel('Time (s)')
plt.suptitle('Drop tower data: subplots', fontsize=18)
plt.subplots_adjust(top=0.9)
#f.savefig('allsubplot.png', dpi=200)
plt.show()
```
